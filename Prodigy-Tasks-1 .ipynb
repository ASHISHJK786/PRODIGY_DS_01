# Visualisation

```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv("Data 1.csv")  
# Make sure the file name and path are correct
```

```python
# If the column names have extra spaces, use df.columns = df.columns.str.strip() to remove them
df.columns = df.columns.str.strip()
```

```python
# If the column names have extra spaces, use df.columns = df.columns.str.strip() to remove them
df.columns = df.columns.str.strip()

# Select only the relevant columns
df_filtered = df[['State', 'Males', 'Females', 'Rural Population']]




# Sort the DataFrame by 'Rural Population' in descending order so the largest bar is first
df_sorted = df_filtered.sort_values(by='Rural Population', ascending=True)



# Create bar width and positions for the groups
bar_width = 0.35
index = np.arange(len(df_filtered))

# Create subplots for Rural and Urban populations
fig, ax = mpl.subplots(figsize=(12, 8))

# Plot bars for Rural and Urban populations
rural_bars = ax.bar(index, df_filtered['Rural Population'], bar_width, label='Rural Population', color='blue')

# Set the labels, title, and ticks
ax.set_xlabel('States')
ax.set_ylabel('Population')
ax.set_title('Rural  Population Distribution by State')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(df_filtered['State'], rotation=45, ha="right")

# Save the figure
mpl.savefig('Rural  Population Distribution by State.png', dpi=300)

# Add legend
ax.legend()

# Display the bar chart
mpl.tight_layout()
mpl.show()
```

```python
# Read the data
df = pd.read_csv("Data 1.csv")

# Check for missing values
print(df.isnull().sum())

# Drop rows where 'Capital' or 'Sex Ratio' is missing
df_cleaned = df.dropna(subset=['Capital', 'Sex Ratio'])

# Convert 'Sex Ratio' column to numeric, coercing errors to NaN and then dropping them
df_cleaned['Sex Ratio'] = pd.to_numeric(df_cleaned['Sex Ratio'], errors='coerce')
df_cleaned = df_cleaned.dropna(subset=['Sex Ratio'])

# Create a horizontal bar chart
mpl.figure(figsize=(10, 6))
mpl.barh(df_cleaned['Capital'], df_cleaned['Sex Ratio'], color='skyblue')

# Add labels and title
mpl.xlabel('Sex Ratio')
mpl.ylabel('Capital City')
mpl.title('Sex Ratio by Capital City of States')

mpl.savefig('Sex Ratio by Capital City of States.png', dpi=300)


# Show the plot
mpl.tight_layout()
mpl.show()
```
